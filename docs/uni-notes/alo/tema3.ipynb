##### 1. Verificare posibilitate descompunere LU

```python
import numpy as np

def descompunere_LU(A, n, eps) :
    for p in range(n):             #calculul elementelor liniei p din matricea U(superior triunghiulara)
        for i in range(p, n) :
            suma = 0
            for k in range(p):
                suma += A[p,k] * A[k,i]
            A[p,i] = A[p,i] - suma
        
        if abs(A[p, p])>eps :      #calculul elementelor coloanei p din matricea L (inferior triunghiulara )
            for i in range(p + 1,n) :
                suma = 0
                for k in range(p) :
                    suma += A[i,k] * A[k, p]
                A[i, p] = (A[i,p] - suma) / A[p, p]
        else:
            print("Nu se poate face impartirea\n")
            return False
    return True

n = 3
eps = 1e-8

A_init = np.array([
    [2.5, 2.0, 2.0],
    [5.0, 6.0, 5.0],
    [5.0, 6.0, 6.5]
], dtype=float)

A = A_init.copy()

if descompunere_LU(A, n, eps) :
    print(A)
```

##### 2. Calcularea determinantului matricei A 

```python
def detA(A, n):            # detA = detL * detU , dar detL = 1 (inferior triunghiulara cu elem de pe diagonala 1 )
    det_A = 1.0            # deci detA = detU ( produsul elem de pe diagonala lui A )
    for i in range(n):
        det_A *= A[i, i]
    return det_A

print("Determinantul matricei A este :" , detA(A,n))
```

##### 3. Calcularea solutiei aproximative a sistemului Ax=b

```python
def rez_sist(A,b, n):
    y = np.zeros(n)
    for i in range(n):
        suma = 0
        for j in range(i) :
            suma += A[i,j] * y[j]
        y[i] = b[i] - suma

    x_LU = np.zeros(n)
    for i in range(n - 1, -1, -1) :
        suma = 0
        for j in range(i + 1, n) :
            suma += A[i, j] * x_LU[j]
        x_LU[i] = (y[i] -suma) /A[i, i]
        
    return x_LU

n = 3
b_init = np.array([2.0, 2.0, 2.0], dtype=float)
x_LU = rez_sist(A, b_init, n)
print("Solutia este :",x_LU)
```

##### 4. Calculul normei 

```python

def norma(A_init, x_LU, b_init, n):
    y_final = np.zeros(n)
    for i in range(n):
        suma = 0
        for j in range(n):
            suma += A_init[i, j] * x_LU[j]
        y_final[i] = suma

    z = np.zeros(n)
    for i in range(n):
        z[i] = y_final[i] - b_init[i]

    suma_patrate = 0
    for i in range(n):
        suma_patrate += z[i]**2
    
    norma_2 = np.sqrt(suma_patrate)
    return norma_2

norma_2 = norma(A_init, x_LU, b_init, n)
print(f"{norma_2:.20f}")
```

##### 5. Compararea cu Numpy

```python
x_lib = np.linalg.solve(A_init, b_init) # Solutia sistemului

A_lib_inv = np.linalg.inv(A_init) # Inversa matricei

x_inv_lib = np.dot(A_lib_inv, b_init) # Inmultirea intre matrice si vector
```

```python
norma_1 = np.linalg.norm(x_LU - x_lib)
norma_2 = np.linalg.norm(x_LU - x_inv_lib)

print(f"Norma ||x_LU - x_lib||: {norma_1}")
print(f"Norma ||x_LU - A_inv_lib * b_init||: {norma_2}")
```

##### 6. Testarea pe dimensiuni mari

```python
n = 100
eps = 1e-10

A_test = np.random.rand(n, n)
b_test = np.random.rand(n)

A_init_test = A_test.copy()

if descompunere_LU(A_test, n, eps):
    x_LU_test = rez_sist(A_test, b_test, n)
    print(f"Eroare sistem: {norma(A_init_test, x_LU_test, b_test, n)}")
```
