##### Functia principala F

```python
import math

# valoarea h este pasul folosit pentru aproximarea derivatelor
# cu cat e mai mic cu atat rezultatul e mai precis
h = 1e-5 # 10^-5

def F(x):
    return (1/3) * (x**3) - 2 * (x**2) + 2 * x + 3
def F2(x):
    return x**2 + math.sin(x)

def F3(x):
    return x**4 - 6 * (x**3) + 13 * (x**2) - 12 * x + 4
```

##### Aproximarile derivatei

```python
def G1(x, h):
    """
        Aproximare folosind punctul curent si doua puncte din spate (x-h, x-2h)
        Este mai rapid dar mai putin precisa decat G2
    """
    return (3*F(x) - 4*F(x - h) + F(x - 2 * h)) / (2 * h)


def G2(x, h):
    """
        Aproximare folosind 2 puncte in stanga si in dreapta lui x (x+2h, x+h, x-h, x-2h)
        Este mult mai precisa decat G1
    """
    return (-F(x + 2 * h) + 8*F(x + h) - 8*F(x - h) + F(x - 2 * h)) / (12 * h)
```

##### Derivata a doua

```python
def F_secund(x, h):
    """
        Aproximarea derivatei de ordin 2
    """
    return (-F(x + 2*h) + 16*F(x + h) - 30*F(x) + 16*F(x - h) - F(x - 2*h)) / (12 * (h**2))
```

- Metoda secantei cere functia g(x). Tema cere sa se ruleze algoritmul de 2 ori (o data cu G1, apoi cu G2) si sa se compare cate iteratii au fost necesare pentru fiecare sa ajunga la aceeasi solutie.
- Metoda secantei gaseste doar un punct critic, adica unde panta e 0, ca sa verificam ca e minim trebuie sa verificam semnul derivatei a doua (>0 e minim)
- Mai este de facut do..while-ul de la pag 3 alegand 2 valori x si x0 si inlocuind delta x cu formula de la pag 2

##### Metoda secantei - aproximare punct minim/maxim/inflexiune --- G1 

```python
import random 

h = 1e-5
eps = 1e-6
k_max = 1000

def generare_random_interval(a, b):
    x0 = random.uniform(a, b)
    x1 = random.uniform(a, b)
    while abs(x1 - x0) < 1e-7: 
        x1 = random.uniform(a, b)
    return x0, x1

x0 , x1 = generare_random_interval(3,4)

k = 0 

while True :
    # delta_x = (( x1 - x0 ) * G1(x1 , h )) / (G1(x1 , h ) - G1(x0 , h))
    numitor = G1(x1 , h ) - G1(x0 , h)
    if numitor >= -eps and numitor <= eps :
        if G1(x1 , h) <= eps/100 :
            delta_x = 0
        else:
            delta_x = 1e-5
    else:
        delta_x = ((x1 - x0)*G1(x1 , h)) / numitor 
    
    x0 = x1 
    x1 = x1 - delta_x
    k = k+1

    if abs(delta_x) < eps or k >= k_max or abs(delta_x) > 1e8 :
        break

if abs(delta_x) < eps:
    print(f"Solutie gasita: x* = {x1} in {k} iteratii")
    
    f2 = F_secund(x1, h)
    
    if f2 > 0:
        print(f"F''(x*) = {f2:.6f} > 0 => Punct MINIM")
    elif f2 < 0:
        print(f"F''(x*) = {f2:.6f} < 0 => Punct MAXIM")
    else:
        print(f"F''(x*) = {f2:.6f} => Punct de inflexiune")
else:
    print("Divergenta")

x_G1 = x1 
k_G1 = k
```

##### Metoda secantei - aproximare punct minim/maxim/inflexiune --- G2

```python
import random 

h = 1e-5
eps = 1e-6
k_max = 1000

def generare_random_interval(a, b):
    x0 = random.uniform(a, b)
    x1 = random.uniform(a, b)
    while abs(x1 - x0) < 1e-7: 
        x1 = random.uniform(a, b)
    return x0, x1

x0 , x1 = generare_random_interval(3,4)

k = 0 

while True :
    delta_x = (( x1 - x0 ) * G2(x1 , h )) / (G2(x1 , h ) - G2(x0 , h))
    numitor = G2(x1 , h ) - G2(x0 , h)
    if numitor >= -eps and numitor <= eps :
        if G2(x1 , h) <= eps/100 :
            delta_x = 0
        else:
            delta_x = 1e-5
    else:
        delta_x = ((x1 - x0)*G2(x1 , h)) / numitor 
    
    x0 = x1 
    x1 = x1 - delta_x
    k = k+1

    if abs(delta_x) < eps or k >= k_max or abs(delta_x) > 1e8 :
        break

if abs(delta_x) < eps:
    print(f"Solutie gasita: x* = {x1} in {k} iteratii")
    
    f2 = F_secund(x1, h)
    
    if f2 > 0:
        print(f"F''(x*) = {f2:.6f} > 0 => Punct MINIM")
    elif f2 < 0:
        print(f"F''(x*) = {f2:.6f} < 0 => Punct MAXIM")
    else:
        print(f"F''(x*) = {f2:.6f} => Punct de inflexiune")
else:
    print("Divergenta")

x_G2 = x1
k_G2 = k
```
