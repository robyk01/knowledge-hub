##### Functia principala F

```python
import math

# valoarea h este pasul folosit pentru aproximarea derivatelor
# cu cat e mai mic cu atat rezultatul e mai precis
h = 1e-5 # 10^-5

def F(x):
    return (1/3) * (x**3) - 2 * (x**2) + 2 * x + 3
def F2(x):
    return x**2 + math.sin(x)

def F3(x):
    return x**4 - 6 * (x**3) + 13 * (x**2) - 12 * x + 4
```

##### Aproximarile derivatei

```python
def G1(x, h):
    """
        Aproximare folosing punctul curent si doua puncte din spate (x-h, x-2h)
        Este mai rapid dar mai putin precisa decat G2
    """
    return 3*F(x) - 4*F(x - h) + F(x - 2 * h) / (2 * h)


def G2(x, h):
    """
        Aproximare folosind 2 puncte in stanga si in dreapta lui x (x+2h, x+h, x-h, x-2h)
        Este mult mai precisa decat G1
    """
    return -F(x + 2 * h) + 8*F(x + h) - 8*F(x - h) + F(x - 2 * h) / (12 * h)
```

##### Derivata a doua

```python
def F_secund(x, h):
    """
        Aproximarea derivatei de ordin 2
    """
    return -F(x + 2*h) + 16*F(x + h) - 30*F(x) + 16*F(x - h) - F(x - 2*h) / (12 * (h**2))
```

- Metoda secantei cere functia g(x). Tema cere sa se ruleze algoritmul de 2 ori (o data cu G1, apoi cu G2) si sa se compare cate iteratii au fost necesare pentru fiecare sa ajunga la aceeasi solutie.
- Metoda secantei gaseste doar un punct critic, adica unde panta e 0, ca sa verificam ca e minim trebuie sa verificam semnul derivatei a doua (>0 e minim)
