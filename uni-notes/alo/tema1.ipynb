### Exercitiul 3: Optimizare

```python
# coeficienti
C1 = 0.333333333333333333 # 1/3
C2 = 0.133333333333333333 # 2/15
C3 = 0.053968253968254 # 17/315
C4 = 0.0218694885361552 # 62/2835

x = float(input())

def my_tan(x):
    x = x % math.pi # periodicitatea
    if x > math.pi / 2:
        x = x - math.pi # anti simetria

    # tan(-x) = -tan(x)
    semn = 1
    if x < 0:
        semn = -1
        x = abs(x) 

    # reducem la (0, pi/4]
    redus = False
    if x > math.pi / 4:
        x = (math.pi / 2) - x
        redus = True

    # calcul polinom - x + x^3 * P(x^2)
    x2 = x * x
    x3 = x * x2

    # P(x^2) = c1 + c2*x^2 + c3*x^4 + c4*x^6
    p_x2 = C1 + x2 * (C2 + x2 * (C3 + x2 * C4)) # schema lui Horner pentru viteza procesor

    tan_aprox = x + x3 * p_x2

    # aplicam tan = 1/tan(pi/2 - x)
    if redus:
        tan_aprox = 1 / tan_aprox if tan_aprox != 0 else float("inf")

    return semn * tan_aprox

print ("Math tan: ", math.tan(x))
print ("My tan: ", my_tan(x), "\n" )


suma = 0 
epsilon = 1e-10

for i in range (1 , 10000 ):
    rand = random.uniform(-math.pi/2 , math.pi/2)
    if ( abs ( rand  - math.pi/2 ) < epsilon or abs ( rand + math.pi/2 ) < epsilon ): # cazuri critice care trebuie evitate (tan tinde la infinit)
        continue 
    suma = suma + abs ( math.tan(rand) - my_tan(rand) )
    # suma erorilor dintre functia my_tan si math.tan

print ("Suma erorilor: ", suma , "\n" )

sum_time3 = 0
# comparam eficienta functiei my_tan cu math.tan 
for i in range (1 , 10000 ):
    rand = random.uniform(-math.pi/2 , math.pi/2)
    start = time.time()
    math.tan(rand)
    end = time.time() 
    sum_time3 = sum_time3 + ( end - start )

print ( "Timp math.tan: " , sum_time3)

sum_time4 = 0
for i in range (1 , 10000 ):
    rand = random.uniform(-math.pi/2 , math.pi/2)
    start = time.time()
    my_tan(rand)
    end = time.time() 
    sum_time4 = sum_time4 + ( end - start )

print ( "Timp my_tan optimizat: " , sum_time4 , "\n" )
```
